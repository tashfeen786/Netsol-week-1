### Basic data types

#### Numbers

Integers and floats work as you would expect from other languages:

````python
x = 3
print(type(x)) # Prints "<class 'int'>"
print(x)       # Prints "3"
````

````python
print(x + 1)   # Addition; prints "4"
print(x - 1)   # Subtraction; prints "2"
print(x * 2)   # Multiplication; prints "6"
print(x ** 2)  # Exponentiation; prints "9"
````

````python
x += 1
print(x)  # Prints "4"
x *= 2
print(x)  # Prints "8"
````

````python
y = 2.5
print(type(y)) # Prints "<class 'float'>"
print(y, y + 1, y * 2, y ** 2) # Prints "2.5 3.5 5.0 6.25"
````

#### Booleans

Python implements all of the usual operators for Boolean logic, but uses English words (`and`, `or`) rather than symbols (`&&`, `||`, etc.):

````python
t = True
f = False
print(type(t)) # Prints "<class 'bool'>"
print(t and f) # Logical AND; prints "False"
print(t or f)  # Logical OR; prints "True"
print(not t)   # Logical NOT; prints "False"
print(t != f)  # Logical XOR; prints "True"
````

#### Strings

Python has great support for strings:

````python
hello = 'hello'    # String literals can use single quotes
world = "world"    # or double quotes; it does not matter.
print(hello)       # Prints "hello"
print(len(hello))  # String length; prints "5"
hw = hello + ' ' + world  # String concatenation
print(hw)  # prints "hello world"
````

````python
# Empty set
empty_set = set()
print(type(empty_set))  # <class 'set'>

# Set with initial values
fruits = {'apple', 'banana', 'orange'}
print(fruits)  # {'banana', 'orange', 'apple'}

# Creating set from a list
numbers = set([1, 2, 3, 2, 1])  # Duplicates removed
print(numbers)  # {1, 2, 3}
````

````python
colors = {'red', 'green', 'blue'}

# Unordered (no index access)


# No duplicates
unique_nums = {1, 2, 2, 3, 3, 3}
print(unique_nums)  # {1, 2, 3}

# Mutable (can add/remove items)
colors.add('yellow')
print(colors)  # {'red', 'green', 'blue', 'yellow'}

# Length of set
print(len(colors))  # 4
````

````python
s = {1, 2, 3}

# Adding elements
s.add(4)
print(s)  # {1, 2, 3, 4}

# Removing elements
s.remove(2)  # Raises KeyError if element doesn't exist
s.discard(5)  # No error if element doesn't exist
print(s)  # {1, 3, 4}

# Pop random element
popped = s.pop()
print(f"Popped: {popped}, Remaining: {s}")

# Clear all elements
s.clear()
print(s)  # set()

# Copy set
new_set = {1, 2, 3}.copy()
````


### 1. Loading and Exploring the Iris Dataset


````python
import pandas as pd
from sklearn.datasets import load_iris

# Load the iris dataset
iris = load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['species'] = iris.target_names[iris.target]

# Display basic information
print("First 5 rows:")
display(iris_df.head())

print("\nLast 5 rows:")
display(iris_df.tail())

print("\nDataset info:")
iris_df.info()

print("\nBasic statistics:")
display(iris_df.describe())
````

### 2. Selecting Data

````python
# Select a single column
sepal_length = iris_df['sepal length (cm)']
print("Sepal length column:")
display(sepal_length.head())

# Select multiple columns
sepal_data = iris_df[['sepal length (cm)', 'sepal width (cm)']]
print("\nSepal measurements:")
display(sepal_data.head())

# Select rows by position
first_10 = iris_df.iloc[:10]
print("\nFirst 10 rows:")
display(first_10)

# Select rows by condition
setosa = iris_df[iris_df['species'] == 'setosa']
print("\nSetosa samples:")
display(setosa.head())
````

### 3. Data Manipulation

````python
# Add a new calculated column
iris_df['sepal_ratio'] = iris_df['sepal length (cm)'] / iris_df['sepal width (cm)']
print("DataFrame with new sepal ratio column:")
display(iris_df.head())

# Group by species
grouped = iris_df.groupby('species')
print("\nMean values by species:")
display(grouped.mean())

# Sort by petal length
sorted_df = iris_df.sort_values('petal length (cm)', ascending=False)
print("\nTop 5 flowers by petal length:")
display(sorted_df.head())
````
